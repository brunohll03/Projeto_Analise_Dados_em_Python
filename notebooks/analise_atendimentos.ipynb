```python
df[df["idade"] < 0]  # Existem registros com idade menor que zero?
```

```python
df[(df["idade"] < 0) | (df["idade"] > 100)]  # Existem registros com idade inválida, menor que 0 ou maior que 100?
```

```python
df.loc[(df["idade"] < 0) | (df["idade"] > 100), "idade"] = pd.NA  # Substitui as idades inválidas por valores ausentes
```

```python
df["cidade"].unique()  # Quais são os diferentes valores de cidade existentes na nossa base?
```

```python
df["cidade"] = df["cidade"].str.strip()  # Remove espaços extras no início e no final dos nomes das cidades
```

```python
df["cidade"] = df["cidade"].str.lower()  # Padroniza os nomes das cidades para letras minúsculas
```

```python
df["especialidade"] = df["especialidade"].str.strip()  # Remove espaços extras no início e no final das especialidades
```

```python
df["especialidade"] = df["especialidade"].str.lower()  # Padroniza os nomes das especialidades para letras minúsculas
```

```python
df["especialidade"].unique()  # Quais são os diferentes valores de especialidade existentes na nossa base?
```

```python
df["status"] = df["status"].str.strip()  # Remove espaços extras no início e no final dos status
```

```python
df["status"] = df["status"].str.lower()  # Padroniza os valores de status para letras minúsculas
```

```python
df["status"].unique()  # Quais são os diferentes valores de status existentes na nossa base?
```

```python
df[df["email"].duplicated(keep=False)]  # Quais registros possuem emails duplicados na nossa base?
```

```python
df[df["telefone"].duplicated(keep=False)]  # Quais registros possuem telefones duplicados na nossa base?
```

```python
df = df.drop_duplicates(subset="email")  # Remove registros duplicados considerando o email como identificador
```

```python
df.drop_duplicates(subset="email", keep="first")  # Remove duplicados pelo email mantendo apenas o primeiro registro
```

```python
df.groupby("cidade")["idade"].mean()  # Qual é a idade média dos pacientes em cada cidade?
```

```python
df.groupby("cidade")["idade"].agg(["count", "mean", "min", "max"])  # Qual é o resumo das idades dos pacientes em cada cidade?
```

```python
df.groupby("cidade")["idade"].agg(["count", "mean", "min", "max"])  # Qual é o resumo das idades dos pacientes em cada cidade?
```

```python
df["idade"].agg([
    "sum",       # Qual é a soma de todas as idades registradas?
    "mean",      # Qual é a idade média dos pacientes?
    "median",    # Qual é a idade que divide os pacientes ao meio?
    "min",       # Qual é a menor idade registrada?
    "max",       # Qual é a maior idade registrada?
    "std",       # Qual é o nível de variação das idades em relação à média?
    "var"        # Qual é a variância das idades registradas?
])  # Quais são as principais estatísticas da distribuição das idades?
```

```python
df["idade"].agg([
    "count",      # Quantos pacientes possuem idade registrada?
    "mean",       # Qual é a idade média?
    "median",     # Qual é a idade central?
    "min",        # Qual é a menor idade?
    "max",        # Qual é a maior idade?
    "std"         # Quanto as idades variam em relação à média?
])  # Qual é o resumo principal das idades dos pacientes?
```

```python
df.groupby("cidade")["idade"].mean().sort_values()  # Qual cidade possui a menor idade média dos pacientes?
```

```python
df.groupby("cidade")["idade"].mean().sort_values(ascending=False)  # Qual cidade possui a maior idade média dos pacientes?
```

```python
df.sort_values("idade")  # Quais são os registros ordenados da menor para a maior idade?
```

```python
df.sort_values("idade", ascending=False)  # Quais são os registros ordenados da maior para a menor idade?
```

```python
df.sort_values(["cidade", "idade"])  # Como os pacientes estão ordenados por cidade e depois por idade?
```

```python
df.sort_values(["cidade", "idade"], ascending=[True, False])  # Como ordenar as cidades em ordem crescente e as idades em ordem decrescente?
```

```python
df.groupby("cidade")["idade"].agg(["count", "mean", "min", "max"]).sort_values("mean", ascending=False)  # Quais cidades possuem as maiores idades médias dos pacientes?
```

```python
df.loc[(df["cidade"] == "campinas") | (df["cidade"] == "vinhedo")]  # Quais pacientes são de Campinas ou Vinhedo?
```

```python
df.loc[(df["idade"] < 0) | (df["idade"] > 100), "idade"] = pd.NA  # Substitui idades inválidas por valores ausentes
```

```python
df["faixa_etaria"] = "adulto"  # Define inicialmente todos os pacientes como adultos
```

```python
df.loc[df["idade"] < 18, "faixa_etaria"] = "menor de idade"  # Classifica pacientes menores de 18 anos
```

```python
df.loc[df["idade"] >= 60, "faixa_etaria"] = "idoso"  # Classifica pacientes com 60 anos ou mais
```
